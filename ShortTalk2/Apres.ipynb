## Procedural Models

```python
from mobilechelonian import Turtle 

def tree(branchLen,t):
    
    if branchLen > 5:
        t.forward(branchLen)
        t.right(40)
        tree(branchLen-10,t)
        t.left(80)
        tree(branchLen-10,t)
        t.right(40)
        t.backward(branchLen)

def main():
    
    t = Turtle()
    t.speed(10)
    t.left(90)
    t.backward(70)
    tree(70,t)

main()
```
